## Create LSTM Sequences

### Subtask:
Generate input sequences (X) and corresponding target sequences (y) from the scaled data, suitable for an LSTM model. This will involve creating sliding windows based on LSTM_INPUT_WINDOW and LSTM_PRED_HORIZON defined in src/config.py.


**Reasoning**:
There is a need to define a function to create sliding window sequences for the LSTM model, considering the specified input features, window size, and prediction horizon, and then apply it to both training and testing datasets.



```python
import numpy as np

def create_sequences(dataset, input_window, pred_horizon):
    X, y = [], []
    # Select the relevant features for the LSTM model
    # 'kWh.mean_value_scaled' and the time-based features
    features = dataset[['kWh.mean_value_scaled', 'year', 'month', 'day', 'hour', 'day_of_week']].values

    for i in range(len(features) - input_window - pred_horizon + 1):
        # Input sequence (X) will be 'input_window' steps long
        X.append(features[i:(i + input_window)])
        # Target (y) will be the 'kWh.mean_value_scaled' 'pred_horizon' steps after the input window
        y.append(features[i + input_window + pred_horizon - 1, 0]) # Assuming target is the scaled kWh value

    return np.array(X), np.array(y)

# Get LSTM configuration from the kernel state
input_window_size = LSTM_INPUT_WINDOW # defined in src/config.py
prediction_horizon = LSTM_PRED_HORIZON # defined in src/config.py

# Generate sequences for training data
X_train, y_train = create_sequences(train_data, input_window_size, prediction_horizon)

# Generate sequences for testing data
X_test, y_test = create_sequences(test_data, input_window_size, prediction_horizon)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

## Define LSTM Model

### Subtask:
Define the LSTM model architecture, taking into account `LSTM_HIDDEN_SIZE`, `LSTM_NUM_LAYERS`, and `LSTM_DROPOUT` from the configuration. This step will set up the neural network structure.


**Reasoning**:
There is a need to import the necessary PyTorch modules and define the LSTM model class as per the instructions. The model will incorporate the LSTM configuration parameters (hidden size, number of layers, dropout) and take into account the input features from X_train.



```python
import torch
import torch.nn as nn

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, dropout):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_size, 1) # Output a single prediction

    def forward(self, x):
        # Initialize hidden and cell state with zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # Pass input through LSTM layer
        out, _ = self.lstm(x, (h0, c0))

        # Take the output from the last time step
        out = self.fc(out[:, -1, :])
        return out

# Determine the input size from X_train's shape
input_features = X_train.shape[2]

# Instantiate the model with parameters from the config
lstm_model = LSTMModel(input_features, LSTM_HIDDEN_SIZE, LSTM_NUM_LAYERS, LSTM_DROPOUT)

print("LSTM Model Architecture:")
print(lstm_model)
```

## Train Model

### Subtask:
Train the defined LSTM model using the training sequences. This involves setting up a training loop, specifying a loss function (e.g., MSE) and an optimizer (e.g., Adam), and training for LSTM_NUM_EPOCHS with the specified LSTM_LEARNING_RATE and LSTM_BATCH_SIZE.


**Reasoning**:
There is a need to convert the NumPy arrays to PyTorch tensors, create TensorDatasets and DataLoaders, define the loss function and optimizer, and then implement the training loop as instructed.



```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

# 1. Convert NumPy arrays to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1) # Unsqueeze for a single output feature
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)

# 2. Create TensorDataset objects
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# 3. Create DataLoader objects
train_loader = DataLoader(train_dataset, batch_size=LSTM_BATCH_SIZE, shuffle=False) # Shuffle can be False for time series
test_loader = DataLoader(test_dataset, batch_size=LSTM_BATCH_SIZE, shuffle=False)

# 4. Define the loss function
criterion = nn.MSELoss()

# 5. Define the optimizer
optimizer = torch.optim.Adam(lstm_model.parameters(), lr=LSTM_LEARNING_RATE)

# Determine device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
lstm_model.to(device)

# 6. Implement the training loop
print(f"Training LSTM model on {device}...")
for epoch in range(LSTM_NUM_EPOCHS):
    lstm_model.train() # Set model to training mode
    total_loss = 0
    for batch_idx, (inputs, targets) in enumerate(train_loader):
        inputs, targets = inputs.to(device), targets.to(device)

        # Forward pass
        outputs = lstm_model(inputs)
        loss = criterion(outputs, targets)

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    print(f'Epoch [{epoch+1}/{LSTM_NUM_EPOCHS}], Loss: {avg_loss:.4f}')

print("Training complete.")
```

**Reasoning**:
Now that the model has been trained, it is needed to evaluate its performance on the test dataset as instructed. This involves making predictions, inverse transforming them, and calculating relevant metrics like MSE and RMSE.



```python
from sklearn.metrics import mean_squared_error
import numpy as np

# 1. Set the model to evaluation mode
lstm_model.eval()

# Lists to store predictions and actual values
predictions = []
actuals = []

# 2. Disable gradient calculations
with torch.no_grad():
    # 3. Iterate through the test_loader to get predictions
    for inputs, targets in test_loader:
        inputs = inputs.to(device)
        outputs = lstm_model(inputs)

        predictions.extend(outputs.cpu().numpy())
        actuals.extend(targets.cpu().numpy())

# Convert lists to numpy arrays
predictions = np.array(predictions).flatten()
actuals = np.array(actuals).flatten()

# 4. Inverse transform the scaled predictions and actual values
# We need a dummy array to inverse transform only the 'kWh.mean_value_scaled' column
# The scaler was fitted on a dataframe with 'kWh.mean_value_scaled' as the first column.
# So, we need to recreate the structure for inverse_transform.

# Create dummy arrays with correct shape for inverse transform
# The scaler was fitted on train_data[['kWh.mean_value']]
# But our features array for creating sequences had 6 columns.
# To inverse transform, the scaler needs an array of shape (n_samples, n_features_in_original_fit).
# The scaler was originally fitted on a single column ('kWh.mean_value').
# So, we should only inverse transform predictions/actuals that were scaled.
# Let's assume the scaler was fit only on the kWh.mean_value column, so it expects a 1-D array or (N,1) array.

# Inverse transform predictions
inverse_predictions = scaler.inverse_transform(predictions.reshape(-1, 1))

# Inverse transform actuals
inverse_actuals = scaler.inverse_transform(actuals.reshape(-1, 1))

# Flatten them back to 1D arrays if needed for metric calculation
inverse_predictions = inverse_predictions.flatten()
inverse_actuals = inverse_actuals.flatten()

# 5. Calculate Mean Squared Error (MSE) and Root Mean Squared Error (RMSE)
mse = mean_squared_error(inverse_actuals, inverse_predictions)
rmse = np.sqrt(mse)

# 6. Print the calculated MSE and RMSE
print(f"Mean Squared Error (MSE) on test data: {mse:.4f}")
print(f"Root Mean Squared Error (RMSE) on test data: {rmse:.4f}")
```

```python
%%writefile src/preprocessing.py
import pandas as pd
import numpy as np
import os
import sys

def clean_and_preprocess_data(df):
    """
    Performs data cleaning and preprocessing steps on the input DataFrame.
    - Converts 'time' column to datetime.
    - Handles missing values in 'kWh.mean_value' by filling with the mean.
    - Removes duplicate rows.
    - Sets 'time' as the DataFrame index.
    - Extracts time-based features (year, month, day, hour, day_of_week).

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: The cleaned and preprocessed DataFrame.
    """
    # Convert 'time' column to datetime format
    df['time'] = pd.to_datetime(df['time'])

    # Handle missing values in 'kWh.mean_value' by filling with the mean
    if 'kWh.mean_value' in df.columns and df['kWh.mean_value'].isnull().any():
        mean_kwh = df['kWh.mean_value'].mean()
        df['kWh.mean_value'] = df['kWh.mean_value'].fillna(mean_kwh)
        print(f"Missing values in 'kWh.mean_value' filled with mean: {mean_kwh:.4f}.")

    # Remove duplicate rows
    initial_rows = len(df)
    df.drop_duplicates(inplace=True)
    final_rows = len(df)
    if initial_rows > final_rows:
        print(f"Removed {initial_rows - final_rows} duplicate rows.")
    else:
        print("No duplicate rows found.")

    # Set 'time' as the DataFrame index
    df = df.set_index('time')

    # Extract time-based features
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    df['hour'] = df.index.hour
    df['day_of_week'] = df.index.dayofweek

    print("Data cleaning and preprocessing complete.")
    return df

if __name__ == '__main__':
    # Ensure the current directory is in the Python path to find 'src'
    if os.getcwd() not in sys.path:
        sys.path.insert(0, os.getcwd())

    # Example usage: Load the raw data using the data_loader and then preprocess
    from src.data_loader import load_raw_data

    example_data_path = 'data/processed/Dataset-of-IoT-Based-Energy-and-Environmental-Parameters-in-a-Smart-Building-Infrastructure-2.0/KWh/Total/2024-06-21-16-47 Chronograf Data Coffee Machine.csv'

    try:
        # Load raw data
        df_raw = load_raw_data(example_data_path)
        print("\nRaw data head before preprocessing:")
        print(df_raw.head())

        # Preprocess data
        df_processed = clean_and_preprocess_data(df_raw.copy())
        print("\nProcessed data head after cleaning and feature engineering:")
        print(df_processed.head())
        print("\nProcessed data info:")
        print(df_processed.info())
    except FileNotFoundError as e:
        print(e)
    except Exception as e:
        print(f"An error occurred during example data preprocessing: {e}")
```

```python
%%writefile src/features.py
import numpy as np
import os
import sys
import pandas as pd

def create_sequences(dataset, input_window, pred_horizon, feature_cols=None):
    """
    Generates input sequences (X) and corresponding target sequences (y)
    from the dataset for an LSTM model using a sliding window approach.

    Args:
        dataset (pd.DataFrame): The input DataFrame containing features.
        input_window (int): The number of time steps for the input sequence (look-back window).
        pred_horizon (int): The number of time steps to predict ahead.
        feature_cols (list, optional): List of column names to use as features.
                                       Defaults to None, which uses all columns except 'kWh.mean_value_scaled' for target.

    Returns:
        tuple: A tuple containing two numpy arrays (X, y),
               where X is the input sequences and y is the target sequences.
    """
    X, y = [], []

    # If feature_cols is not provided, assume all columns in dataset except the target are features
    if feature_cols is None:
        # Assuming 'kWh.mean_value_scaled' is always the first column and target
        features = dataset.values
    else:
        features = dataset[feature_cols].values

    target_column_index = dataset.columns.get_loc('kWh.mean_value_scaled')

    for i in range(len(features) - input_window - pred_horizon + 1):
        # Input sequence (X) will be 'input_window' steps long
        X.append(features[i:(i + input_window)])
        # Target (y) will be the 'kWh.mean_value_scaled' 'pred_horizon' steps after the input window
        y.append(features[i + input_window + pred_horizon - 1, target_column_index])

    return np.array(X), np.array(y)

if __name__ == '__main__':
    # Ensure the current directory is in the Python path to find 'src'
    if os.getcwd() not in sys.path:
        sys.path.insert(0, os.getcwd())

    # Example usage (assuming train_data, LSTM_INPUT_WINDOW, LSTM_PRED_HORIZON are defined)
    # For this example, we'll create dummy data resembling the actual notebook state.

    # Dummy data similar to df_coffee_machine after preprocessing and scaling
    data = {
        'kWh.mean_value_scaled': np.random.rand(100), # Scaled values
        'year': np.full(100, 2024),
        'month': np.random.randint(1, 13, 100),
        'day': np.random.randint(1, 29, 100),
        'hour': np.random.randint(0, 24, 100),
        'day_of_week': np.random.randint(0, 7, 100)
    }
    dummy_df = pd.DataFrame(data, index=pd.to_datetime(pd.date_range('2024-01-01', periods=100, freq='H')))

    # Define parameters (example values from config.py)
    LSTM_INPUT_WINDOW = 24
    LSTM_PRED_HORIZON = 1

    # Define feature columns to be used for the LSTM input
    feature_cols = ['kWh.mean_value_scaled', 'year', 'month', 'day', 'hour', 'day_of_week']

    print("\n--- Running example from src/features.py ---")
    X_dummy, y_dummy = create_sequences(dummy_df, LSTM_INPUT_WINDOW, LSTM_PRED_HORIZON, feature_cols=feature_cols)

    print(f"Shape of dummy X: {X_dummy.shape}")
    print(f"Shape of dummy y: {y_dummy.shape}")
    print("Dummy sequences created successfully.")
```

This file (`src/models_lstm.py`) will contain the definition of the LSTM (Long Short-Term Memory) neural network model. Encapsulating the model architecture here promotes modularity and makes the model easily reusable and importable in other parts of the project.

```python
%%writefile src/models_lstm.py
import torch
import torch.nn as nn
import os
import sys

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, dropout):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_size, 1) # Output a single prediction

    def forward(self, x):
        # Initialize hidden and cell state with zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # Pass input through LSTM layer
        out, _ = self.lstm(x, (h0, c0))

        # Take the output from the last time step
        out = self.fc(out[:, -1, :])
        return out

if __name__ == '__main__':
    # Ensure the current directory is in the Python path to find 'src'
    if os.getcwd() not in sys.path:
        sys.path.insert(0, os.getcwd())

    # Dummy values for demonstration, similar to config.py
    # In a real script, these would be imported from config.
    DUMMY_INPUT_FEATURES = 6 # Example, should match X_train.shape[2]
    LSTM_HIDDEN_SIZE = 64
    LSTM_NUM_LAYERS = 2
    LSTM_DROPOUT = 0.2

    # Instantiate the model with dummy parameters
    lstm_model = LSTMModel(DUMMY_INPUT_FEATURES, LSTM_HIDDEN_SIZE, LSTM_NUM_LAYERS, LSTM_DROPOUT)

    print("LSTM Model Architecture (example):")
    print(lstm_model)

    # Example forward pass with dummy data
    # Dummy input: (batch_size, sequence_length, input_features)
    dummy_input = torch.randn(10, 24, DUMMY_INPUT_FEATURES) # 10 samples, 24 timesteps, 6 features
    dummy_output = lstm_model(dummy_input)
    print(f"\nShape of dummy output: {dummy_output.shape}")
```

This file (`src/models_ml.py`) will contain implementations of various traditional machine learning models for time series forecasting, such as `RandomForestRegressor`.

```python
%%writefile src/models_ml.py
import numpy as np
import pandas as pd
import os
import sys
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix
from sklearn.preprocessing import MinMaxScaler

# Ensure the current directory is in the Python path to find 'src'
if os.getcwd() not in sys.path:
    sys.sys.path.insert(0, os.getcwd())

# Assuming these are available from config or passed as arguments
# For standalone execution, these might be defined here or imported from a config file
LSTM_INPUT_WINDOW = 24
LSTM_PRED_HORIZON = 1

class MLModelTrainer:
    def __init__(self, model_type='RandomForestRegressor', **model_params):
        if model_type == 'RandomForestRegressor':
            self.model = RandomForestRegressor(**model_params)
        else:
            raise ValueError(f"Unsupported model type: {model_type}")
        self.scaler = None

    def prepare_data(self, df, target_column='kWh.mean_value', input_window=LSTM_INPUT_WINDOW, pred_horizon=LSTM_PRED_HORIZON):
        # Scale the target column
        self.scaler = MinMaxScaler()
        df['kWh.mean_value_scaled'] = self.scaler.fit_transform(df[[target_column]])

        # Create sequences for ML models (flatten input window into features)
        X, y = [], []
        features = df[['kWh.mean_value_scaled', 'year', 'month', 'day', 'hour', 'day_of_week']].values

        for i in range(len(features) - input_window - pred_horizon + 1):
            X.append(features[i:(i + input_window)].flatten()) # Flatten window into a single feature vector
            y.append(features[i + input_window + pred_horizon - 1, 0]) # Target is the scaled kWh value

        return np.array(X), np.array(y)

    def train(self, X_train, y_train):
        print(f"Training {type(self.model).__name__}...")
        self.model.fit(X_train, y_train)
        print("Training complete.")

    def predict(self, X_test):
        return self.model.predict(X_test)

    def evaluate(self, y_true, y_pred):
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_true, y_pred)
        print(f"Mean Squared Error (MSE): {mse:.4f}")
        print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
        print(f"R-squared (R2): {r2:.4f}")
        return mse, rmse, r2

def train_occupancy_classifier(df, feature_cols, target_col, test_size=0.2, random_state=42):
    """
    Trains an occupancy classifier (RandomForestClassifier) and returns the model, scaler, and metrics.
    """
    # Prepare data
    X = df[feature_cols]
    y = df[target_col]

    # Scale features
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    X_scaled_df = pd.DataFrame(X_scaled, columns=feature_cols, index=df.index)

    # Split data chronologically
    split_point = int(len(X_scaled_df) * (1 - test_size))
    X_train, X_test = X_scaled_df.iloc[:split_point], X_scaled_df.iloc[split_point:]
    y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

    # Initialize and train model
    model = RandomForestClassifier(n_estimators=100, random_state=random_state)
    model.fit(X_train, y_train)

    # Evaluate
    y_pred = model.predict(X_test)

    report = classification_report(y_test, y_pred, output_dict=True)
    cm = confusion_matrix(y_test, y_pred).tolist() # Convert to list for JSON serialization

    metrics = {
        "classification_report": classification_report(y_test, y_pred),
        "confusion_matrix": cm
    }

    return model, scaler, metrics

if __name__ == '__main__':
    # Example usage for RandomForestRegressor
    print("\n--- Running example from src/models_ml.py (RandomForestRegressor) ---")

    # Create a dummy DataFrame similar to df_coffee_machine for demonstration
    data = {
        'kWh.mean_value': np.random.rand(500) * 50 + 10, # Unscaled values
        'year': np.full(500, 2024),
        'month': np.random.randint(1, 7, 500),
        'day': np.random.randint(1, 29, 500),
        'hour': np.random.randint(0, 24, 500),
        'day_of_week': np.random.randint(0, 7, 500)
    }
    dummy_df = pd.DataFrame(data, index=pd.to_datetime(pd.date_range('2024-01-01', periods=500, freq='h')))

    # Initialize ML model trainer
    ml_trainer = MLModelTrainer(model_type='RandomForestRegressor', n_estimators=100, random_state=42)

    # Prepare data (scaling and sequence creation)
    X, y = ml_trainer.prepare_data(dummy_df.copy())

    # Split data into train and test sets (chronological split)
    split_point = int(len(X) * 0.8)
    X_train_ml, y_train_ml = X[:split_point], y[:split_point]
    X_test_ml, y_test_ml = X[split_point:], y[split_point:]

    print(f"Shape of X_train for ML: {X_train_ml.shape}")
    print(f"Shape of y_train for ML: {y_train_ml.shape}")
    print(f"Shape of X_test for ML: {X_test_ml.shape}")
    print(f"Shape of y_test for ML: {y_test_ml.shape}")

    # Train model
    ml_trainer.train(X_train_ml, y_train_ml)

    # Make predictions
    predictions_ml_scaled = ml_trainer.predict(X_test_ml)

    # Inverse transform predictions and actuals
    predictions_ml = ml_trainer.scaler.inverse_transform(predictions_ml_scaled.reshape(-1, 1)).flatten()
    actuals_ml = ml_trainer.scaler.inverse_transform(y_test_ml.reshape(-1, 1)).flatten()

    # Evaluate model
    print("\nModel Evaluation on Test Data (Inverse Transformed):")
    ml_trainer.evaluate(actuals_ml, predictions_ml)

    print("Example RandomForestRegressor run complete.")
```

```python
%%writefile src/train_lstm.py
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
import os
import sys

# Ensure the current directory is in the Python path to find 'src'
if os.getcwd() not in sys.path:
    sys.path.insert(0, os.getcwd())

from src.models_lstm import LSTMModel

# Assuming these are available globally or imported from a config file
# For standalone execution, we'll use dummy values similar to what's in config.py
LSTM_INPUT_WINDOW = 24
LSTM_PRED_HORIZON = 1
LSTM_HIDDEN_SIZE = 64
LSTM_NUM_LAYERS = 2
LSTM_DROPOUT = 0.2
LSTM_BATCH_SIZE = 64
LSTM_NUM_EPOCHS = 30
LSTM_LEARNING_RATE = 1e-3

def train_lstm_model(
    X_train,
    y_train,
    X_test,
    y_test,
    input_features,
    hidden_size=LSTM_HIDDEN_SIZE,
    num_layers=LSTM_NUM_LAYERS,
    dropout=LSTM_DROPOUT,
    batch_size=LSTM_BATCH_SIZE,
    num_epochs=LSTM_NUM_EPOCHS,
    learning_rate=LSTM_LEARNING_RATE
):
    """
    Trains an LSTM model for time series forecasting.

    Args:
        X_train (np.ndarray): Training input sequences.
        y_train (np.ndarray): Training target values.
        X_test (np.ndarray): Testing input sequences.
        y_test (np.ndarray): Testing target values.
        input_features (int): Number of input features per time step.
        hidden_size (int): Number of hidden units in the LSTM layer.
        num_layers (int): Number of recurrent layers.
        dropout (float): Dropout probability.
        batch_size (int): Batch size for training.
        num_epochs (int): Number of training epochs.
        learning_rate (float): Learning rate for the optimizer.

    Returns:
        torch.nn.Module: The trained LSTM model.
    """
    # 1. Convert NumPy arrays to PyTorch tensors
    X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
    X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)

    # 2. Create TensorDataset objects
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

    # 3. Create DataLoader objects
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    # 4. Instantiate the model
    model = LSTMModel(input_features, hidden_size, num_layers, dropout)

    # 5. Define the loss function and optimizer
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    # Determine device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # 6. Implement the training loop
    print(f"Training LSTM model on {device}...")
    for epoch in range(num_epochs):
        model.train()  # Set model to training mode
        total_loss = 0
        for batch_idx, (inputs, targets) in enumerate(train_loader):
            inputs, targets = inputs.to(device), targets.to(device)

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, targets)

            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        avg_loss = total_loss / len(train_loader)
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')

    print("Training complete.")
    return model, test_loader, device

if __name__ == '__main__':
    # Example usage (assuming X_train, y_train, X_test, y_test are defined from previous steps)
    print("\n--- Running example from src/train_lstm.py ---")
    # For this example, we'll use the already defined X_train, y_train, X_test, y_test from the notebook state.
    # These were generated in the 'Create LSTM Sequences' subtask.

    if 'X_train' in locals() and 'y_train' in locals() and 'X_test' in locals() and 'y_test' in locals():
        input_features = X_train.shape[2]

        # Train the model
        trained_model, test_loader, device = train_lstm_model(
            X_train, y_train, X_test, y_test, input_features
        )

        print("\nSuccessfully trained LSTM model using the example data.")
        print(f"Trained model device: {device}")

    else:
        print("Error: X_train, y_train, X_test, or y_test not found in the current environment.")
        print("Please ensure the 'Create LSTM Sequences' step has been run.")
```
